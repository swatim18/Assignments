**8. Why is eggs a valid variable name while 100 is invalid?**

Ans 8:

A variable name can have A-Z, a-z, 0-9 and an underscore(\_), but it
cannot start with a number. Hence ‘eggs’ is a valid variable name as it

only a number, and a variable cannot be only a number or start with a
number.

**10. Why does this expression cause an error? How can you fix it?**

**'I have eaten ' + 99 + ' burritos.'**

Ans 10:

**'I have eaten ' + 99 + ' burritos.'** This statement causes an error
as we are concatenating two different data types. Quotation marks
include string type value however 99 is integer thus concatenation
without typecasting is causing the error.

We can fix it as follows:

**'I have eaten ' + str(99) + ' burritos.'**
